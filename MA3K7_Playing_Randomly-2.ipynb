```python
import numpy as np
import matplotlib.pyplot as plt
def game_random_fill(n):
    # For even n, ensure equal number of 0s and 1s; for odd n, one more of either 0 or 1
    total_elements = n * n
    if n % 2 == 0:  # Even n
        ones_count = total_elements // 2
    else:  # Odd n
        ones_count = total_elements // 2 + np.random.choice([0, 1])
    
    zeros_count = total_elements - ones_count
    
    # Create a flat array with the correct counts of 0s and 1s
    flat_matrix = np.array([1]*ones_count + [0]*zeros_count)
    
    # Shuffle the array to randomize 0 and 1 placements
    np.random.shuffle(flat_matrix)
    
    # Reshape the flat array back into an n x n matrix
    matrix = flat_matrix.reshape(n, n)
    
    # Calculate the determinant of the matrix
    det = np.linalg.det(matrix)
    
    return matrix, det
```

```python
# Prepare data for plotting
n_values = list(range(1, 31))
probabilities = []

for n in n_values:
    determinant_zero_count_random = 0
    k = 10
    for i in range(k):
        i, det_random = game_random_fill(n)
        # print(i)
        if det_random == 0:
            determinant_zero_count_random += 1
    probability_determinant_is_zero1 = float((determinant_zero_count_random / k)*100)
    probability_str = "{:.10f}".format(probability_determinant_is_zero1)
    probabilities.append(probability_determinant_is_zero1)
    print('probability of determinant equals to 0 for n =', n, ':', probability_str, '%')
    
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, probabilities, marker='o', linestyle='-', color='b')
plt.title('Probability of Determinant Being Zero vs. Matrix Size n')
plt.xlabel('Matrix Size n')
plt.ylabel('Probability of Determinant Being Zero (%)')
plt.grid(True)
plt.xticks(n_values)
plt.show()
```
